```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

viridis = cm.get_cmap('viridis', 256)
#newcolors = viridis(np.linspace(0, 1, 256))
#pink = np.array([248/256, 24/256, 148/256, 1])
#newcolors[:25, :] = pink
#newcmp = ListedColormap(newcolors)

def linear_map(x):
    return x

#trig blue red
x = np.linspace(0, np.pi/2, 256)
newcolors = np.array([np.sin(x), np.zeros(x.size), np.cos(x), np.ones(x.size)]).T
newcmp = ListedColormap(newcolors)

#linear blue red
x = np.linspace(0, 1, 256)
newcolors = np.array([linear_map(x), np.zeros(x.size), 1 - linear_map(x), np.ones(x.size)]).T
newcmp1 = ListedColormap(newcolors)

#linear white red
x = np.linspace(0, 1, 256)
newcolors = np.array([np.ones(x.size), 1-linear_map(x), 1-linear_map(x), np.ones(x.size)]).T
newcmp2 = ListedColormap(newcolors)

#trig white red
x = np.linspace(0, np.pi/2, 256)
newcolors = np.array([np.ones(x.size), 1-np.sin(x), 1-np.sin(x), np.ones(x.size)]).T
newcmp3 = ListedColormap(newcolors)

#trig green red
x = np.linspace(0, np.pi/2, 256)
newcolors = np.array([np.sin(x), np.cos(x), np.zeros(x.size), np.ones(x.size)]).T
newcmp4 = ListedColormap(newcolors)

#linear green red
x = np.linspace(0, 1, 256)
newcolors = np.array([linear_map(x), 1-linear_map(x), np.zeros(x.size), np.ones(x.size)]).T
newcmp5 = ListedColormap(newcolors)

def plot_examples(cms):
    """
    helper function to plot two colormaps
    """
    np.random.seed(19680801)
    data = np.random.randn(30, 30)

    fig, axs = plt.subplots(1, 7, figsize=(21, 3), constrained_layout=True)
    for [ax, cmap] in zip(axs, cms):
        psm = ax.pcolormesh(data, cmap=cmap, rasterized=True, vmin=-4, vmax=4)
        fig.colorbar(psm, ax=ax)
    plt.show()

plot_examples([viridis, newcmp, newcmp1, newcmp2, newcmp3, newcmp4, newcmp5])
```
